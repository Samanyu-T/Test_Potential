```python
def create_entry(data, n_vac, n_h, n_he, val_type , val, ref):

    defect_key = 'V%dH%dHe%d' % (n_vac, n_h, n_he)

    if defect_key not in data.keys():
        data[defect_key] = {}
        data[defect_key]['eam'] = {}
        data[defect_key]['eam']['val'] = []
        data[defect_key]['eam']['ref'] = []

        data[defect_key]['dft'] = {}
        data[defect_key]['dft']['val'] = []
        data[defect_key]['dft']['ref'] = []

        data[defect_key]['new'] = {}
        data[defect_key]['new']['val'] = []
        data[defect_key]['new']['ref'] = []

        data[defect_key]['r_vol_dft'] = {}
        data[defect_key]['r_vol_dft']['val'] = []
        data[defect_key]['r_vol_dft']['ref'] = []

        data[defect_key]['r_vol_new'] = {}
        data[defect_key]['r_vol_new']['val'] = []
        data[defect_key]['r_vol_new']['ref'] = []

    if val_type not in data[defect_key].keys():
        
        data[defect_key][val_type] = {}
        data[defect_key][val_type]['val'] = []
        data[defect_key][val_type]['ref'] = []
    
    data[defect_key][val_type]['val'].append(val)
    data[defect_key][val_type]['ref'].append(ref)

    return data

def find_formation_from_binding(data, defect1, defect2, val_type, ref, binding):

    defect1_key = 'V%dH%dHe%d' % (defect1[0], defect1[1], defect1[2])

    defect2_key = 'V%dH%dHe%d' % (defect2[0], defect2[1], defect2[2])

    final_defect = np.array(defect1) + np.array(defect2)

    ref_array1 = np.array(data[defect1_key][val_type]['ref'])

    ref_array2 = np.array(data[defect2_key][val_type]['ref'])

    idx1 = np.where(ref_array1 == ref)

    idx2 = np.where(ref_array2 == ref)

    if len(idx1[0]) > 0:
        defect1_energy = data[defect1_key][val_type]['val'][idx1[0][0]]
    else:
        defect1_energy = np.mean(np.array(data[defect1_key][val_type]['val']))

    if len(idx2[0]) > 0:
        defect2_energy = data[defect2_key][val_type]['val'][idx2[0][0]]
    else:
        defect2_energy = np.mean(np.array(data[defect2_key][val_type]['val']))
    
    final_defect_energy = defect1_energy + defect2_energy - binding

    data = create_entry(data, final_defect[0], final_defect[1], final_defect[2], val_type, final_defect_energy, ref)

    return data

```
