```python
def He_edensity(r):

    sigma = 0.75
    rho = (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-r**2/(2*sigma**2))
    
    return 4*rho

def He_eam(rho):

    F = 1.5*np.sqrt(rho) + 0.25

    return F
```

```python
with open('Potentials/Tungsten_Hydrogen_Helium/WHHe_He_edensity.eam.alloy','w') as file:

    file.write('# W-H-He potential \n')
    file.write('# \n')
    file.write('# \n')

    file.write('3 W H He \n')
    
    with open('Potentials/Tungsten_Hydrogen_Helium/WHHe_test.eam.alloy', 'r') as ref:
    

        for i in range(4):
            ref.readline()
        

        line = ref.readline()
        file.write(line)

        val = line.split()
        Nrho = int(val[0])
        drho = float(val[1])
        Nr   = int(val[2])
        dr   = float(val[3])
        cutoff  = float(val[4])
        
        
        line = ref.readline()
        file.write(line)

# F(rho) rho(r) Tungsten
        for i in range(Nrho):

            line = ref.readline()
            file.write(line)


        for i in range(Nr):

            line = ref.readline()
            file.write(line)

        line = ref.readline()
        file.write(line)

# F(rho) rho(r) Hydrogen
        for i in range(Nrho):

            line = ref.readline()
            file.write(line)


        for i in range(Nr):

            line = ref.readline()
            file.write(line)

        
        line = ref.readline()
        file.write(line)

# F(rho) rho(r) Helium
        for i in range(Nrho):

            line = ref.readline()
            rho = drho*i
            val = He_eam(rho)
            
            file.write('%16.8f \n' % val)

        for i in range(Nr):

            line = ref.readline()
            r = drho*i
            val = He_edensity(r)
            
            file.write('%16.8f \n' % val)

# Pairwise W-W, W-H, H-H, W-He, H-He, He-He
        for i in range(3*Nr):
            line = ref.readline()
            file.write(line)
        
        for i in range(Nr):
            line = ref.readline()
            val  = float(line)
            val = 0.5*val
            file.write('%16.8f \n' % val)

        for i in range(2*Nr):
            line = ref.readline()
            file.write(line)
        
```
